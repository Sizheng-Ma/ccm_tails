```python
def normalization(q):
    eta = q/(1+q)**2
    return eta
```

```python
q1 = np.loadtxt('q1/Psi4_truncated.dat')
q2 = np.loadtxt('q2/Psi4_truncated.dat')
q4 = np.loadtxt('q4/Psi4_truncated.dat')
```

# Comparisons

```python
pl.semilogy(q1[:,0], abs(q1[:,1])/normalization(1), lw=2, label=r'$q=1$')
pl.semilogy(q2[:,0], abs(q2[:,1])/normalization(2), lw=2, label=r'$q=2$')
pl.semilogy(q4[:,0]+351.5, abs(q4[:,1])/normalization(4), lw=2, label=r'$q=4$')


pl.xlim(1800, 2600)
pl.ylim(1e-10, 1e-1)
pl.legend()
pl.xlabel('$u/M$')
pl.ylabel(r'$\Psi_4^{\ell=2,m=0}/\eta$')
```
